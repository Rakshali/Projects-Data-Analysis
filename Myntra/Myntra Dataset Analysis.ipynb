2. Data Cleaning

2.1) Removing extra columns

```python
del df['img_link']
del df['product_tag']
del df['brand_tag']
```

```python
df
```

2.2) Creating other required columns by splitting the product_link (product_tag, brand_tag, product_id)

(Removed columns product_tag and brand_tag so extract more appropriate values under these columns fro product link)

```python
df_productlink = df['product_link'].reset_index()
df_productlink['product_link'] = df['product_link'].str.split('/')
```

```python
df['product_tag'] = df_productlink['product_link'].apply(lambda x: x[0])
df['brand_tag'] = df_productlink['product_link'].apply(lambda x: x[1])
df['product_id'] = df_productlink['product_link'].apply(lambda x: x[3])
```

```python
df
```

2.3) Checking if all values of rating_count have non-negative values

```python
count = 0
for i in df.rating_count:
  if i >= 0:
    count+=1
  else:
    print(df.rating_count)
print(count)
```

```python
df['discount_amount'] = df['marked_price'] - df['discounted_price']
df
```

```python
df['discount_percentage'] = (df['discount_amount']/df['marked_price']) * 100
df['discount_percentage'] = df['discount_percentage'].astype(int)
df
```

Checking if discounted_price of the products is less than marked_price or not

```python
count = 0
for i in df.discount_amount:
  if i >= 0:
    count += 1
  else:
    print(df.product_id)
print(count)
```

Checking if the discount_percentage (0 to 100) of all the products is valid or not

```python
count = 0
for i in df['discount_percentage']:
  if (i >= 0) and (i <= 100):
    count += 1
  else:
    print('product_id')
print(count)
```

3.14) Distribuition of Ratings

```python
bins = [0,2,3,4,5]
df_ratings = pd.cut(df['rating'], bins=bins, labels=['< 2','2 - 3','3 - 4','4 - 5'])
df_ratings_grouped = df_ratings.value_counts().reset_index()
df_ratings_grouped.columns = ['ratings','count']
df_ratings_grouped
```

```python
df_ratings_grouped.plot(kind='bar', title = 'Distribution of ratings',x = 'ratings', y = 'count',ylabel = 'count')
```

```python
labels=['4 - 5','3 - 4','2 - 3', '< 2']
df_ratings_grouped.plot(kind='pie', title = 'Distribution of ratings',y = 'count',autopct = '%1.1f%%',radius = 0.65,labeldistance=1.5,pctdistance = 1.2,textprops={'fontsize': 7}, labels = labels)
```
